#**Logical** **Operators**

1. logical and **and**

2. Logical or. **or**

3. Logical not. **not**

#**Logical** **and**
 It returns **true** if both statements are **true**, **false** otherwise

```python
#st1.      st2.      result

#true.     true.     true
#true.     false     False
#false     true.     False
#false     false     false
```

```python
10 > 5 and 20 > 10 #true and True
```

```python
10 > 5 and 30 > 89 #true and false
```

```python
30 > 67 and 10 > 5 #false and true
```

```python
30 > 56 and 10 > 20 #false and false
```

#**Logical** **or**

It returns **True** if one of the statement is **True**

```python
#st1.    st2.    Result

#True. True.  True
#True. False. True
#False. True. True
#False False False
```

```python
10 > 5 or 20 > 10 #true or True
```

```python
10 > 5 or 30 > 89 #True or False
```

```python
30 > 67 or 10 > 85 #False or True
```

```python
30 > 56 or 10 > 20  #false or false
```
